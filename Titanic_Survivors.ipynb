HANDELING NULL VALUES

The dataset may contain many rows and columns for which some values are missing,we can't leave those missing value as it is.
 1.either drop the entire row or column
 2.fill missing values with some appropriate value(eg.mean of particular row or column)


```python
df.isnull().sum()
```

separating out the columns which have more than 35% of the values missing in the dataset.

```python
drop_col=df.isnull().sum()[df.isnull().sum()>35/100*df.shape[0]]
drop_col
drop_col.index
```

```python
df.drop(drop_col.index,axis=1,inplace=True)
df.isnull().sum()
```

```python
df.fillna(df.mean(),inplace=True)
df.isnull().sum()
```

because embarked contains string values,we could not calculate mean

```python
df['Embarked'].describe()
```

for embarked attribute,we fill null values with the most repeated value in the column

```python
df['Embarked'].fillna('S',inplace=True)
df.isnull().sum()
```

proportionality between values

```python
df.corr()
```

make new column family_size by combining sibsp and parch columns

```python
df['family_size']=df['SibSp']+df['Parch']
df.drop(['SibSp','Parch'],axis=1, inplace=True)
df.corr()
```

```python
df['Alone']=[0 if df['family_size'][i]>0 else 1 for i in df.index]
df.head()
```

```python
df.groupby(['Alone'])['Survived'].mean()
```

```python
df[['Alone','Fare']].corr()
```

```python
df['Sex']=[0 if df['Sex'][i]=='male' else 1 for i in df.index]
df.groupby(['Sex'])['Survived'].mean()
```

```python
df.groupby(['Embarked'])['Survived'].mean()
```
